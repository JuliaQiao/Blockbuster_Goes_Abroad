```python
import pandas as pd

df1 = pd.read_pickle('mojo_df_100.pickle')
df2 = pd.read_pickle('mojo_df_200.pickle')
df3 = pd.read_pickle('mojo_df_300.pickle')
df4 = pd.read_pickle('mojo_df_400.pickle')
df5 = pd.read_pickle('mojo_df_500.pickle')
df6 = pd.read_pickle('mojo_df_600.pickle')
df7 = pd.read_pickle('mojo_df_700.pickle')
df8 = pd.read_pickle('mojo_df_800.pickle')
df9 = pd.read_pickle('mojo_df_900.pickle')
df10 = pd.read_pickle('mojo_df_1000.pickle')
df11 = pd.read_pickle('mojo_df_1100.pickle')
df12 = pd.read_pickle('mojo_df_1200.pickle')
```

```python
dfs = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12]
```

```python
mojo_compiled_df = pd.concat(dfs)
```
