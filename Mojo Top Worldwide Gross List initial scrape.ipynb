```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.boxofficemojo.com/chart/ww_top_lifetime_gross/'

response = requests.get(url)
```

```python
response.status_code
```

```python
response.text[1000:2000]
```

```python
page = response.text
```

```python
soup = BeautifulSoup(page)
```

```python
table = soup.find(id='table')
```

```python
print(table).prettify()
#class="a-text-left mojo-field-type-title" 
#class="a-link-normal" hr
```

```python
rows = [row for row in table.find_all('tr')]
```

```python
rows[1]
```

```python
rows[1].find('a')['href']
#https://www.boxofficemojo.com/
```

```python
title = rows[1].find('a')
title
```

```python
title.text
```

```python
#class="a-link-normal" href="/title/tt4154796/?ref_=bo_cso_table_1">Avengers: Endgame
```

```python
movies = {}

for row in rows[1:]:
    items = row.find_all('td')
    link = items[1].find('a')
    title, url = link.text, link['href']
    movies[title] = [url] + [i.text for i in items]
    
movies
```

```python
top_movies = pd.DataFrame(movies).T  #transpose
top_movies.columns = ['Link', 'Rank', 'Title', 'World_Gross', 
                    'Domestic_Gross', 'Domestic_Percentage', 'Foreign_Gross', 'Foreign_Percentage','Year']
top_movies.tail()
```
