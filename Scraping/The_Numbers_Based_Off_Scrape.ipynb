```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def Get_Titles(urls):
    """
    Concatenates scraped titles of The Numbers' "All Time Domestic Based on Fiction " table.
    Input: list of partial URLs referring to pages of table. 
    Output: concatenated list of title column of table across all inputed pages. 
    """ 
    movies = []
    
    page_1 = 'https://www.the-numbers.com/box-office-records/domestic/all-movies/sources/based-on-fictional-book-or-short-story/{}'
    
    for url in (urls):
        url = page_1.format(url)
        response = requests.get(url)
        page = response.text
        soup = BeautifulSoup(page)
        table = soup.find('table')
        rows = [row for row in table.find_all('tr')]
        for row in rows[1:]:
            items = row.find_all('td')
            link = items[2].find('a')
            title= link.text
            movies.append(title)
    return movies
    
    
```

```python
urls = [0, 101, 201, 301, 401, 501, 601, 701, 801, 901, 1001, 
        1101, 1201, 1301, 1401, 1501, 1601, 1701, 1801, 1901, 2001]
books = Get_Titles(urls)
```
